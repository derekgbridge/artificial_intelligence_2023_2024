<h1>Recommender Systems</h1>
<ul>
    <li><b>Recommender systems</b> suggest which items (movies, music, new stories, &hellip;) a user should consume.</li>
    <li>Ideally, the suggestions are:
        <ul>
            <li><b>personalised</b>: the recommender system tailors its suggestions based on what it knows about
                the user:
                <ul>
                    <li>the user’s tastes, interests, preferences;</li>
                    <li>her personality;</li>
                    <li>her long-term goals; and</li>
                    <li>her skills, knowledge.</li>
                </ul>
            </li>
            <li><b>contextualised</b>: the recommender system tailors its suggestions to the circumstances:
                <ul>
                    <li>the time;</li>
                    <li>the place (physical or virtual);</li>
                    <li>the conditions (weather, etc.);</li>
                    <li>the user’s companions;</li>
                    <li>her mood; and</li>
                    <li>her short-term goals.</li>
                </ul>
            </li>
        </ul>
    </li>
</ul>

<h2>Why Recommender Systems?</h2>
<ul>
   <li>Recommender Systems are a response to information overload: users want to discover items from among
       an overwhelming body of items.
       <ul>
           <li>Recommender systems do not make choices for the user.</li>
           <li>Instead, recommender systems help the user to manage choice.</li>
       </ul>
    </li>
    <li>Question: Why not browse?</li>
    <li>Question: Why not search?</li>
    <li>Recommender systems are an AI success story: 
        <ul>
            <li>They are used by Netflix, Spotify, Amazon, Google News, Twitter, LinkedIn, &hellip; They are one of the few examples of AI <em>at scale</em>.
            </li>
            <li>A McKinsey report from 2013 stated that “35 percent of what consumers purchase on Amazon and 75 percent of what they watch on Netflix come from product recommendations based on such algorithms”.
            </li>
    </li>
    <li>They're not without problems. Think Facebook! Privacy, filter bubbles, polarization, inappropriate content,
        &hellip; Maybe we are more aware of these problems precisely because recommender systems have become so
        widely used.
    </li>  
</ul>

<h2>The Objective</h2>
<ul>
    <li>To <b>create joy</b>!</li>
    <li>Less idealistically, businesses have objectives such as: to increase revenue, to increase sales, to increase profit, to increase the number of customers, to retain
        customers, to increase repeat visits, and so on.
        <figure>
            <img src="images/rs_purposes.png" alt="Goals of a RS" />
            <figcaption>From: Jannach, D., & Bauer, C. (2020). Escaping the McNamara Fallacy: Towards more Impactful Recommender Systems Research. AI Magazine, 41(4), 79-95. https://doi.org/10.1609/aimag.v41i4.5312</figcaption>
        </figure>
    </li>
    <li>But, in the early years, we used AI to predict ratings,
        <ul>
            <li>e.g. given a user and an item, we predict how many
                'stars' (1-5) this user would give to this item.
            </li>
        </ul>
        In this case, our objective is to minimize prediction error. Note how different this is from the objectives
        listed above. This difference between the business objectives and the objective of the AI system is something
        we have discussed before. In a later lecture, we will discuss broader objectives &mdash; ones that go beyond 
        just minimizing prediction error.
    </li>
</ul>

<h2>What data might we have about items?</h2>
<ul>
    <li>Let's refer to the set of all items as $I$ and an individual item as $i$. The number
        of items is $|I|$.
    </li>
    <li>We can represent each item $i \in I$ as a vector.</li>
    <li>What are the features?
        <ul>
            <li>For products, we might have features such as size, weight, manufacturer, etc.
                Similarly, for movies, we might have director, length, language, guidance certificate, etc.; for songs, it may be artist, producer, record label, etc.
            </li>
            <li>We might have categories, defined by a domain expert, e.g. movie genres (comedy, thriller, romance, &hellip;), song genres (rock, country, hip-hop, &hellip;). These can become vectors of binary-valued or numeric-valued features.
            </li>
            <li>We might have keywords, e.g. automatically extracted from a description of the item, such as from a movie synopsis. These might become vectors of binary-valued or numeric-valued features (as we saw with bag-of-words representations).</li>
        </ul>
    </li>
</ul>

<h1>Users</h1>
<ul>
    <li>Let's refer to the set of all users as $U$ and an individual user as $u$. The number
        of users is $|U|$.</li>
    <li>We can represent each user $u \in U$ as a vector.</li>
    <li>What are the features?
        <ul>
            <li>In systems where users must create an account, we can obtain the values of features during the sign-up
                process, e.g.:
                <ul>
                    <li>demographic features, such as sex, age, level of education;</li>
                    <li>interests, maybe as categories or as keywords.
                    </li>
                </ul>
            </li>
        </ul>
    </li>
</ul>

<h2>Sparsity</h2>
<ul>
    <li>The sparsity of an interaction matrix is the proportion of its entries that are $\bot$.
        <ul>
            <li>Again, if you like maths: $1 - \frac{|\v{\Omega}|}{|U|\times|I|}$
    </li>
    <li>In most commercial scenarios,
        <ul>
            <li>we have a very large number of items, e.g. thousands, even millions;</li>
            <li>many users, e.g. thousands or tens of thousands;</li>
            <li>even the most active user is likely to have rated $< 1\%$;</li>
            <li>hence, the matrix is very sparse;
            <li>sparsity of $> 90\%$ or even $> 99\%$ is common.</li>
        </ul>
    </li>
    <li>Suppose we generalize from an interaction matrix (rank 2 tensor) to a rank 3 tensor or higher so that
        we can record contextual information:
        <ul>
            <li>e.g. an extra dimension to record on which day-of-the-week the rating was made;</li>
            <li>e.g. an extra dimension to record whether it was sunny, cloudy or raining when the rating was made;</li>
            <li>etc.
        </ul>
        Then sparsity is even greater.
    </li>
    <li>Sparsity is one of the biggest challenges when building recommender systems.
        <ul>
            <li>E.g. it makes it hard to find similar rows (users) or columns (items) or to find meaningful patterns in the data.</li>
        </ul>
    </li>
</ul>

<h2>Cold-start items</h2>
<ul>
    <li>When a new item becomes available, it will have no ratings.</li>
    <li>For a long time, it may have few ratings.
    </li>
    <li>Some recommender systems algorithms are better at coping with this situation than others.</li>
    <li>There is even a possible catch-22:
        <ul>
            <li>items that aren't rated don't get recommended;</li>
            <li>items that don't get recommended don't get rated.</li>
       </ul>
    </li>
    <li>In some domains, new items are the most common. Give an example!</li>
</ul>

<h2>Cold-start users</h2>
<ul>
    <li>New users will have no ratings (unless we require them to rate some items during sign-up).</li>
    <li>Infrequent users may remain in cold-start.</li>
    <li>Some recomender systems algorithms are better at coping with this situation than others.</li>
    <li>In some domains, all or most users are anonymous and hence in cold-start. Give an example!</li>
</ul>

<ul>
    <li>Recommender systems typically proceed through (at least) three steps:
        <figure>
            <img src="images/rs_arch.png" />
        </figure>
    </li>
    <li>Before, we discuss these in a tiny bit more detail, let's make the observation that recomender systems are always domain-specific. What applies in one domain may not apply in another domain.
    </li>
</ul>

<h2>Recommending is domain-specific</h2>
<ul>
    <li>The unit of recommendation:
        <ul>
            <li>individual items, packages, sequences (e.g. playlists, tours).</li>
        </ul>
    </li>
    <li>The target consumer:
        <ul>
            <li>individual users, small groups (e.g. families, housemates), larger groups 
                (occupants of a shared space, communities).
            </li>
        </ul>
    </li>
    <li>Level of interaction:
        <ul>
            <li>passive, confirmation (e.g. skipping a song), selection from a list.</li>
        </ul>
    </li>
    <li>The nature of the item:
        <ul>
            <li>high-value versus low-value;</li>
            <li>high consumption cost versus low consumption cost;</li>
            <li>rivalrous versus non-rivalrous;</li>
            <li>perishable versus non-perishable;</li>
            <li>one-off consumption versus repeated consumption;</li>
            <li>&hellip;and so on!</li>
        </ul>
    </li>
    <li>The nature of the recommendation, e.g.:
        <ul>
            <li>items that could be alternatives to the one the user is viewing;</li>
            <li>items that are complementary to the one the user is viewing;</li>
            <li>items that might come next after consuming the item the user is consuming;</li>
            <li>&hellip;and so on!</li>
        </ul>
    </li>
</ul>

<h2>Candidate generation</h2>
<ul>
    <li>Since there are so many items, we choose a smaller subset of candidate items.</li>
    <li>Examples:
        <ul>
            <li>Candidates for user $u$ might be her un-rated items, i.e. items $i$ for which $r_{ui}=\bot$. This has two potential problems. What are they? <!-- Too big. Assumes never want repeat recommendations. -->
            </li>
            <li>For "linear TV", candidates might be programmes that are being broadcast this evening.</li>
            <li>For movie-going, candidates might be films that are being screened at the user's multiplex this week.</li>
            <li>For online news, candidates will be recent stories.</li>
            <li>For on-the-go travel (e.g. restaurants, hotels and points-of-interest), candidates must be nearby
                and open.
            </li>
        </ul>
    </li>
</ul>

<h2>Scoring</h2>
<ul>
    <li>Each candidate item is scored, e.g. for how relevant it is to this user, allowing the candidates to be ranked in order of decreasing score.
    </li>
    <li>There are many ways of doing this.
        <ul>
            <li>The next part of this lecture shows one way: its scores are similarities between users and items.
            </li>
            <li>The next two lectures give several algorithms that take a supervised learning approach: there, the scores are predicted ratings.
            </li>
        </ul>
    </li>
</ul>

<h2>Top-$N$ recommendation</h2>
<ul>
    <li>The last step is top-$N$ recommendation:
        <ul>
            <li>The obvious thing to do is to select the $N$ candidates whose scores are highest and recomend these to the user.</li>
        </ul>
    </li>
    <li>However, there may be some additional criteria to take into account at this stage. Some examples include:
        <ul>
            <li>There may be some business rules to take into account. For example, there may be some items the business
                is trying to push. So there may be a rule that requires that one or more slots in the top-$N$ are occupied by these items, displacing the 'organic' recommendations. (Think about sponsored content, for example.)
            </li>
            <li>We might have more than one recommender model whose scores we want to combine.</li>
            <li>We often carry out some re-ranking at this stage in order to ensure that the top-$N$ has a 
                degree of diversity or some notion of fairness.
            </li>
        </ul>
    </li>
</ul>

<h1>A Simple Content-Based Recommender</h1>
<ul>
    <li>We'll end this lecture by describing one way of doing scoring.</li>
    <li>This is called a content-based recommender because it uses <em>descriptions</em> of items and users.</li>
    <li>It works by describing users and items using the same features.</li>
    <li>For a given user, the score of a candidate item, is the similarity between the user and the item.</li>
</ul>

<h2>Scores are similarities</h2>
<ul>
    <li>The score that capture the relevance to user $u$ of item $i$ is simply the similarity of vectors $\v{P}^{(u)}$ and $\v{Q}^{(i)}$.
    </li>
    <li>As discussed, previously, we can use cosine similarity for this. And, if we ignore
    normalization, this is simply the product of the two vectors.
        $$sim(u, i) = \v{P}^{(u)}\v{Q}^{(i)}$$
    </li>
    <li>E.g. the scores of all the movies for $u_4$ are:
        $$sim(u_4,i_1)=2; sim(u_4, i_2)=0; sim(u_4, i_3)=1; sim(u_4, i_4)=1; sim(u_4, i_5)=2; sim(u_4, i_6)=0$$
    </li>
    <li>Other formulae are possible!</li>
</ul>

<h2>Discussion</h2>
<ul>
    <li>This requires item and user descriptions in terms of content-based features, such as genres.</li>
    <li>We can imagine that, when a new item (e.g. movie) is released, its producers describe the item by giving
        a list of its genres.
    </li>
    <li>But where will we get the user descriptions?</li>
    <li>Question: how well does this cope with cold-start items?</li>
    <li>Question: how well does this cope with cold-start users?</li>
    <li>It may make relevant recommendations.
        <ul>
            <li>This depends on the quality of these hand-engineered features: if the features are too coarse, for
                example, the recommendations may not seem relevant. 
            </li>
            <li>It also depends on having feature-values that are not too noisy.</li>
        </ul>
    </li>
    <li>It may also be able to explain its recommendations. How?</li>
    <li>One disadvantage is that this is unlikely to extend a person's
        existing interests, i.e. it is unlikely to make surprising recommendations. Why not?
    </li>
</ul>
